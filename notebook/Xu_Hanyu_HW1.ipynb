#### iii. Split Data Set

```python
#split data by class(0 AND 1)
df_ab=df.loc[df['class'] ==b'Abnormal' ]
df_nor=df.loc[df['class'] ==b'Normal' ]

#GET first 70 from normal subgroup and first 140 from abnormal subgroup 
train_nor=df_nor[0:70]
test_nor=df_nor[70:100]
train_abn=df_ab[0:140]
test_abn=df_ab[140:210]

#combine to create train data and test data
train_df=pd.concat([train_abn,train_nor])
test_df=pd.concat([test_abn,test_nor])
```

```python
x_train=train_df[["pelvic_incidence","pelvic_tilt","lumbar_lordosis_angle","sacral_slope","pelvic_radius","degree_spondylolisthesis"]]

y_train=train_df['class']

x_test=test_df[["pelvic_incidence","pelvic_tilt","lumbar_lordosis_angle","sacral_slope","pelvic_radius","degree_spondylolisthesis"]]

y_test=test_df['class']

y_train=y_train.replace(b'Abnormal',1)
y_train=y_train.replace(b'Normal',0)

y_test=y_test.replace(b'Abnormal',1)
y_test=y_test.replace(b'Normal',0)
```

### (c) Classification

#### i. Euclidean Metric

```python
# using the KNeighborsClassifier, in this case I choose k=3 , and default is Euclidean metric.
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
sc=knn.score(x_test,y_test)
sc
```

#### ii. Test Data

```python
k_range = range(208,0,-3)
error_rate= []
min_a=1
best_index=208
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    a=(1-knn.score(x_test, y_test))
    if a <= min_a:
        min_a=a
        best_index=k
    

print("the best k which returens the lowest error rate is:",best_index)
print("with k= :",best_index,"the lowest error rate is:",min(error_rate))

plt.plot(k_range, error_rate)
plt.title("errors in terms of k")
plt.xlabel('k')
plt.ylabel('error rate')
plt.show() 
```

```python
# while k=4
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(x_train,y_train)
sc=knn.score(x_test,y_test)


# predict y-value using test data
y_predict = knn.predict(x_test)

c=y_test.tolist()
y_p=y_predict.tolist()

confu_matrix = confusion_matrix(c, y_p)
print("the confusion matrix is:")
print(confu_matrix)

true_positive_rate = 25/(25+5) #RECALL
true_negetive_rate =69/(69+1)
precision=25/(25+1)
f_value=2*true_positive_rate*precision/(true_positive_rate+precision)
print("the true positive rate is:", true_positive_rate)
print("the true negative rate is:", true_negetive_rate)
print("the precision is :", precision)
print("the f score is :", f_value)
```

#### iii. Learning Curve

```python
n_range = range(10,211,10)
size_n_error=[]
for n in n_range:
    
    new_train_xnormal=x_train[140:(140+int(n/3))]
    new_train_xabnormal=x_train[0:(n-int(n/3))]
    new_train_x=pd.concat([new_train_xabnormal,new_train_xnormal])
    
    new_train_ynormal=y_train[140:(140+int(n/3))]
    new_train_yabnormal=y_train[0:(n-int(n/3))]
    new_train_y=pd.concat([new_train_yabnormal,new_train_ynormal])
    k_range=range(1,n,5)
    error_rate=[]
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors = k)
        knn.fit(new_train_x, new_train_y)
        error_rate.append(1-knn.score(x_test, y_test))
    size_n_error.append(min(error_rate))
        
plt.plot(n_range, size_n_error)  
plt.title("learning curve based on size of N")
plt.xlabel('N training sample size')
plt.ylabel('error rate')
plt.show() 
        
```

### (d) Other Metrics

#### i. Minkowski Distance.

##### A. Manhattan Distance with p = 1.

```python
k_range = range(1,200,5)
error_rate= []
min_a=1
best_index=1
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k,p=1)
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    a=(1-knn.score(x_test, y_test))
    if a <= min_a:
        min_a=a
        best_index=k
print("the best k with the lowest rate is:", best_index,"and it has a error rate:",min(error_rate))
```

##### B. With log10(p) in {0.1, 0.2, 0.3, ... ,1}.

```python
# from the previous question, I choose k=26
p_range=range(1,11,1)
error_rate=[]
p_value=[]
for p in p_range:
    a=p/10
    x=10**(a)
    p_value.append(x)
    knn = KNeighborsClassifier(n_neighbors = 26,p=x)
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    
plt.plot(p_value,error_rate)
plt.xlabel('p')
plt.ylabel('error rate')
plt.title("error rate plot for for log(p)")
plt.show() 

print(error_rate)

print("based on the graph and the list of error rate, the first, the second,the 4th and 6th all have the lowest error rate.")
print("which is when log(p)=0.1/0.2/0.4/0.6")
print("the lowest error rate is:",min(error_rate))
```

##### C. Chebyshev Distance With p -> infinity.

```python
k_range = range(1,200,5)
error_rate= []
min_a=1
best_index=1
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k,p=float('inf'))
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    a=(1-knn.score(x_test, y_test))
    if a <= min_a:
        min_a=a
        best_index=k
print("the most reasonable k when p is infinite is:" ,best_index)
print("when k is:",best_index,"which result as the lowest error rate:",min(error_rate))
```

#### ii. Mahalanobis Distance.

```python
k_range = range(1,200,5)
error_rate= []
min_a=1
best_index=1
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k,metric='mahalanobis',metric_params={'VI': np.linalg.inv(np.cov(x_train.T))})
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    a=(1-knn.score(x_test, y_test))
    if a <= min_a:
        min_a=a
        best_index=k
print("the most reasonable k when metric is mahalanobis:" ,best_index)
print("when k is:",best_index,"which result as the lowest error rate:",min(error_rate))

plt.plot(k_range,error_rate)
plt.xlabel('k')
plt.ylabel('error rate')
plt.show()
```

### (e) Weighted Decision

##### (1). Euclidean Distance with p = 2.

```python
k_range = range(196,0,-5)
error_rate= []
min_a=1
best_index=1
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k,weights='distance',p=2)
    knn.fit(x_train, y_train)
    error_rate.append(1-knn.score(x_test, y_test))
    a=(1-knn.score(x_test, y_test))
    if a <= min_a:
        min_a=a
        best_index=k

print("the most reasonable k when p is 2 with weighted decision is:" ,best_index)
print("when k is:",best_index,"which result as the lowest error rate:",min(error_rate))
plt.plot(k_range,error_rate)
plt.xlabel('k')
plt.ylabel('error rate')
plt.show()
```
